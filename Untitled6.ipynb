```python
#auxilio henry com outra maneira de fazer

dia = input("Digite o dia da semana: ").strip().lower()
dias_uteis = ["segunda", "terça", "quarta", "quinta", "sexta"]

# O resultado da verificação (True ou False) vira a chave do dicionário
status = {
    True: "A pessoa está de descanso",
    False: "A pessoa está em dia útil"
}

descanso = dia not in dias_uteis
print(status[descanso])
```
